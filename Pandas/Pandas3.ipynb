```python
# 행, 열 선택, 삭제 및 필터

import pandas as pd

friends = [
    [25,'student'],
    [28,'developer'],
    [30,'teacher']    
]

df = pd.DataFrame(friends, index=['Sean','Ray','Tim'], columns=['age','job'])
df
```

```python
df.iloc[:,0:2]
```

```python
df.iloc[0:2,0:2]
```

```python
df.drop(['Sean','Tim'])
```

```python
# inplace=True: drop 한 value가 바로 DataFrame에 적용되어 따로 assign 할 필요 없음
df.drop(['Sean','Tim'], inplace=True)
df
```

```python
df = df.drop(df.index[[0,2]])
df
```

```python
df = df.drop('age', axis=1)
df
```

```python
df_book = pd.read_csv("book_data.csv", encoding="cp949")
df_book.head()
```

```python
book_filtered = df_book[['도서명','저자명']]
book_filtered.head()
```

```python
df_book.filter(items=['도서명','저자명']).head()
```

```python
# filter(like=컬럼명에 포함된 값, axis=축)
df_book.filter(like='명', axis=1).head()
```

```python
# filter(regex=정규식)
# '_$': _로 끝나는 것
df_book.filter(regex='명$', axis=1).head()
```
